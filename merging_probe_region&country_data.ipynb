```python
import pandas as pd
import json
import os

# Load probe data with error handling for inconsistent rows
probe_data = pd.read_csv('./probe_data.csv', on_bad_lines='skip')

# Function to parse traceroute JSON data and add probe info
def parse_traceroute_data(traceroute_data_json, probe_data):
    traceroute_records = []
    for record in traceroute_data_json:
        try:
            traceroute_dict = json.loads(record)
            prb_id = traceroute_dict.get("prb_id")
            # Look up country and region based on probe ID
            probe_info = probe_data[probe_data['Probe ID'] == prb_id]
            if not probe_info.empty:
                country = probe_info['Country'].values[0]
                region = probe_info['Region'].values[0]
            else:
                country = None
                region = None
            traceroute_dict['Country'] = country
            traceroute_dict['Region'] = region
            traceroute_records.append(traceroute_dict)
        except json.JSONDecodeError:
            print(f"Error decoding JSON for record: {record}")  # Log any JSON errors
    return traceroute_records

# Function to parse ping JSON data and add probe info
def parse_ping_data(ping_data_json, probe_data):
    ping_records = []
    for record in ping_data_json:
        try:
            ping_dict = json.loads(record)
            prb_id = ping_dict.get("prb_id")
            # Look up country and region based on probe ID
            probe_info = probe_data[probe_data['Probe ID'] == prb_id]
            if not probe_info.empty:
                country = probe_info['Country'].values[0]
                region = probe_info['Region'].values[0]
            else:
                country = None
                region = None
            ping_dict['Country'] = country
            ping_dict['Region'] = region
            ping_records.append(ping_dict)
        except json.JSONDecodeError:
            print(f"Error decoding JSON for record: {record}")  # Log any JSON errors
    return ping_records

# Load JSON data files as text, assuming they contain newline-separated JSON objects
def load_json_file(file_path):
    with open(file_path, 'r') as file:
        return file.readlines()

# Load traceroute and ping data
traceroute_data_json = load_json_file('./16.12.11.4 - AWS/TRACEROUTE-RIPE-Atlas-measurement-80565700-1729407600-to-1729423807.json')
ping_data_json = load_json_file('./16.12.11.4 - AWS/PING-RIPE-Atlas-measurement-80565864-1729407600-to-1729423662.json')

# Parse traceroute and ping data and merge with probe data
traceroute_records = parse_traceroute_data(traceroute_data_json, probe_data)
ping_records = parse_ping_data(ping_data_json, probe_data)

# Save the merged data to JSON files
output_dir = './16.12.11.4 - AWS/'
os.makedirs(output_dir, exist_ok=True)

# Save traceroute records to JSON
with open(os.path.join(output_dir, 'traceroute_data_merged.json'), 'w') as file:
    json.dump(traceroute_records, file)

# Save ping records to JSON
with open(os.path.join(output_dir, 'ping_data_merged.json'), 'w') as file:
    json.dump(ping_records, file)

print("Merged traceroute and ping data saved to JSON files.")
```
